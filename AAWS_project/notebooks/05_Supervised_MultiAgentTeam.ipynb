# 감독형 멀티에이전트 팀 구축

지금까지 우리는 04 실습에서 **Navigator → Coder 파이프라인**을 순서대로 수동 연결했습니다.
이번에는 한 단계 더 나아가, 이 두 에이전트를 **한 명의 감독자(Supervisor)가 지휘하는 팀**으로 재구성합니다.

## 이번 실습에서 배우는 것
| 이전 실습 (04) | 이번 실습 (05) |
|---|---|
| Navigator, Coder를 수동으로 순서대로 호출 | Supervisor가 상황을 판단해 자동 위임 |
| Blueprint를 꺼내 전달 | 에이전트끼리 협업하여 자율 수행 |
| 단일 파이프라인 (순차) | 감독형 팀 (계층적 구조) |

## 팀 구조

어떤 멀티에이전트 패턴을 써야 할까?

멀티 에이전트 패턴에 대한 문서를 참고하여 두 가지 패턴의 차이를 이해하고 이번 실습에 적합한 방식을 선택하세요.

### 패턴 1: Supervisor with Handoffs
- Supervisor가 **핸드오프 도구**로 제어권을 워커에게 이전
- 워커는 전체 대화 맥락을 그대로 받음
- 자유로운 대화 흐름, 협업이 자연스러움

### 패턴 2: Supervisor with Agent as Tool
- Supervisor가 워커를 **도구처럼 호출** (Call & Return)
- 워커에게는 필요한 정보만 선별해서 전달
- **맥락 격리** → 토큰 절약, 환각 방지
- Navigator와 Coder는 서로의 대화 내역을 알 필요가 없습니다.
- Supervisor가 Navigator의 Blueprint 결과만 뽑아서 Coder에게 전달하면 충분합니다.



## 실습 대상: Quotes to Scrape

**http://quotes.toscrape.com**

웹 스크래핑 연습을 위해 만들어진 공개 사이트입니다.


## 시나리오 1: 다중 페이지 수집 ⭐
**목표**: Quotes to Scrape 사이트의 **1~5페이지**에서 모든 인용구 데이터를 수집하세요.
**수집 항목**:
- 인용구 원문 (text)
- 저자 이름 (author)
- 태그 목록 (tags)
**결과 형식** (`quotes_5pages.json`):
```json
[
  {
    "text": "The world as we have created it...",
    "author": "Albert Einstein",
    "tags": ["change", "deep-thoughts", "thinking", "world"]
  },
  ...
]
```

## 시나리오 2: 태그 필터 수집 ⭐⭐
**목표**: 특정 태그(예: book)를 가진 인용구를 전체 페이지에서 모두 수집하세요.

**수집 항목**:
- 인용구 원문 (text)
- 저자 이름 (author)
- 태그 목록 (tags)
**결과 형식** (`quotes_tag_inspirational.json`):

```json
[
  {
    "text": "I have not failed. I've just found 10,000 ways that won't work.",
    "author": "Thomas A. Edison",
    "tags": ["edison", "failure", "inspirational", "paraphrased"]
  },
  ...
]
```

## 시나리오 3: 2단계 분석 수집 ⭐⭐⭐
**목표**: love 태그를 가진 인용구들을 먼저 분석한 뒤, 그 안에서 가장 많이 등장하는 태그 Top 10을 찾고, 각 태그별로 인용구를 수집하세요.

**수행 단계**:


[1단계] love 태그 인용구 전체 수집
        → http://quotes.toscrape.com/tag/love/

[2단계] 수집된 인용구들의 태그를 분석
        → 빈도수 기준 Top 10 태그 추출
        → 예: ["love", "life", "romance", "relationships", ...]

[3단계] Top 10 태그 각각에 대해 인용구 수집
        → http://quotes.toscrape.com/tag/{tag_name}/
        → 태그마다 별도 JSON 파일로 저장


결과 형식:
```
quotes_tag_love.json
quotes_tag_life.json
quotes_tag_romance.json
...
```

## 시나리오 4: 동적 사이트 수집 및 데이터 시각화 (도전 과제) 🌟🌟🌟🌟

지금까지는 주로 정적인 텍스트 데이터를 JSON으로 정리하는 데 그쳤습니다.
이번에는 동적 요소가 포함된 실전 사이트에서 데이터를 수집하고, 결과를 시각화하는 작업까지 에이전트 팀에게 맡겨보세요!

**추천 대상 사이트**:
- [Scrape This Site - Oscar Winning Films](https://scrapethissite.com/pages/ajax-javascript/) : AJAX 동적 렌더링된 연도별 데이터 수집
- [Scrape This Site - Countries of the World](https://scrapethissite.com/pages/simple/) : 국가별 체급(인구수/면적) 수치 데이터
- 그 외 Github Trending(`https://github.com/trending`), 환율/주식 시세 등 원하는 곳 자유 선택

**목표**:
1. Navigator와 Coder를 활용해 대상 사이트의 수치형 데이터(예: 국가별 인구, 연도별 수상작 횟수, 코인 등락률 등)를 추출해 JSON/CSV로 저장합니다.
2. 데이터 수집이 끝나면, 수집된 파일을 읽어 **matplotlib, seaborn 또는 pandas 등을 사용하여 막대 그래프/산점도 차트**를 그리고 `.png` 이미지 파일로 저장하는 코드를 만들어 실행하게 하세요.

**힌트 및 과제 핵심**:
- 이 과제를 수행하려면 Coder(수집 전문) 외에 파이썬 데이터 분석/시각화 코드를 짤 **`Data Analyst` (데이터 분석가) 역할의 에이전트**를 팀에 추가로 구축하는 것이 가장 좋습니다.
- Supervisor가 다음과 같은 흐름으로 업무를 위임하도록 프롬프트를 설계해야 합니다:
  `👉 1. Navigator(사이트 구조 파악) → 2. Coder(크롤러 개발 및 데이터 저장) → 3. Analyst(데이터 읽어서 차트 저장)`
- 분석가 에이전트가 작성하는 코드에는 화면 표출(`plt.show()`)이 아닌 **파일 저장(`plt.savefig('chart.png')`)** 이 반드시 포함되어야 합니다.
