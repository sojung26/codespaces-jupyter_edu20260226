- êµ¬ì¡°ê°€ ê·¼ë³¸ì ìœ¼ë¡œ ë‹¤ë¥¸ ê²½ìš° (ë‹¤ë¥¸ ì‚¬ì´íŠ¸ ë“±) â†’ **Blueprint ë³„ë„ ìƒì„±**

````python
import json
from typing import Optional
from pydantic import BaseModel, Field, field_validator

# ==========================================
# ë™ì  Nê³„ì¸µ Blueprint ìŠ¤í‚¤ë§ˆ
# ==========================================

class PageLayer(BaseModel):
    """í•˜ë‚˜ì˜ íƒìƒ‰ ê³„ì¸µì„ í‘œí˜„í•˜ëŠ” ë‹¨ìœ„ ë¸”ë¡"""
    layer_name: str = Field(
        description="ì´ ê³„ì¸µì˜ ì—­í•  ì´ë¦„ (ì˜ˆ: 'ê¸°ì‚¬ ëª©ë¡', 'ìƒí’ˆ ìƒì„¸')"
    )
    url_pattern: str = Field(
        description="ì´ ê³„ì¸µì˜ URL êµ¬ì¡° ì˜ˆì‹œ ë˜ëŠ” ì§„ì…ì  URL (ì‹¤ì œ ì‹œì‘ URLì€ entry_urls ì°¸ì¡°)"
    )
    selectors: dict[str, str] = Field(
        description="ì´ ê³„ì¸µì—ì„œ ìˆ˜ì§‘í•  ë°ì´í„°ì˜ CSS ì…€ë ‰í„° ë”•ì…”ë„ˆë¦¬ (key: í•„ë“œëª…, value: CSS ì…€ë ‰í„°)"
    )
    navigate_to_next: Optional[str] = Field(
        default=None,
        description="ë‹¤ìŒ ê³„ì¸µìœ¼ë¡œ ì´ë™í•˜ëŠ” ë§í¬ì˜ CSS ì…€ë ‰í„°. ë§ˆì§€ë§‰ ê³„ì¸µì´ë©´ ë°˜ë“œì‹œ None."
    )
    pagination_method: Optional[str] = Field(
        default=None,
        description="í˜ì´ì§€ë„¤ì´ì…˜ ë°©ì‹ (URLíŒŒë¼ë¯¸í„° / AJAXë²„íŠ¼ / ë¬´í•œìŠ¤í¬ë¡¤ / None)"
    )

    @field_validator("selectors", mode="before")
    @classmethod
    def parse_selectors(cls, v):
        if isinstance(v, str):
            try:
                return json.loads(v)
            except Exception:
                pass
        return v

    # â†“ ì´ validator ì¶”ê°€
    @field_validator("navigate_to_next", "pagination_method", mode="before")
    @classmethod
    def parse_none_string(cls, v):
        """LLMì´ Noneì„ ë¬¸ìì—´ "None"ìœ¼ë¡œ ë°˜í™˜í•˜ëŠ” ê²½ìš°ë¥¼ ì²˜ë¦¬í•©ë‹ˆë‹¤."""
        if v in ("None", "null", "ì—†ìŒ", "N/A", ""):
            return None
        return v

class NavigatorBlueprint(BaseModel):
    """Navigatorê°€ Coderì—ê²Œ ì „ë‹¬í•˜ëŠ” ë™ì  Nê³„ì¸µ í¬ë¡¤ë§ ì„¤ê³„ ë„ë©´ (ë‹¨ì¼ êµ¬ì¡°)"""
    entry_urls: list[str] = Field(
        description=(
            "í¬ë¡¤ë§ì„ ì‹œì‘í•  URL ëª©ë¡. "
            "êµ¬ì¡°(ê³„ì¸µ/ì…€ë ‰í„°)ê°€ ë™ì¼í•˜ê³  ì‹œì‘ì ë§Œ ë‹¤ë¥¸ ê²½ìš° ì—¬ëŸ¬ ê°œ ì§€ì •. "
            "ì˜ˆ) ì •ì¹˜ ì„¹ì…˜ URL + ì‚¬íšŒ ì„¹ì…˜ URL"
        )
    )
    total_layers: int = Field(
        description="íƒìƒ‰ì— í•„ìš”í•œ ì´ ê³„ì¸µ ìˆ˜ (layers ë¦¬ìŠ¤íŠ¸ì˜ ê¸¸ì´ì™€ ë™ì¼)"
    )
    layers: list[PageLayer] = Field(
        description="íƒìƒ‰ ìˆœì„œëŒ€ë¡œ ì •ë ¬ëœ PageLayer ëª©ë¡. layers[0]ì€ entry_urls ê°ê°ì— ë°˜ë³µ ì ìš©ë¨."
    )
    rendering_type: str = Field(
        description="Static SSR ë˜ëŠ” Dynamic CSR/JS"
    )
    anti_bot_notes: str = Field(
        description="ë¡œê·¸ì¸ í•„ìš” ì—¬ë¶€, íŒì—…, ìº¡ì°¨, ìš°íšŒ ì¡°ì–¸ ë“±. ì—†ìœ¼ë©´ 'ì—†ìŒ'"
    )


class NavigatorBlueprintCollection(BaseModel):
    """Navigatorê°€ ë°˜í™˜í•˜ëŠ” Blueprint ëª¨ìŒ (1ê°œ ì´ìƒ)"""
    total_jobs: int = Field(
        description="ì´ Blueprint ìˆ˜. êµ¬ì¡°ê°€ ê°™ìœ¼ë©´ 1ê°œ, êµ¬ì¡°ê°€ ë‹¤ë¥¸ ì‚¬ì´íŠ¸/ì„¹ì…˜ì€ ê°ê° 1ê°œ."
    )
    blueprints: list[NavigatorBlueprint] = Field(
        description=(
            "ìˆ˜ì§‘ ì‘ì—…ë³„ Blueprint ëª©ë¡. "
            "- êµ¬ì¡° ë™ì¼ + ì‹œì‘ URLë§Œ ë‹¤ë¦„ â†’ Blueprint 1ê°œ, entry_urlsì— ë³µìˆ˜ URL "
            "- êµ¬ì¡°ê°€ ê·¼ë³¸ì ìœ¼ë¡œ ë‹¤ë¦„ â†’ Blueprintë¥¼ ë³„ë„ ìƒì„±í•˜ì—¬ ë³µìˆ˜ ë°˜í™˜"
        )
    )


# ==========================================
# âœ… ìŠ¤í‚¤ë§ˆ ê²€ì¦ ì˜ˆì‹œ
# ==========================================

# ì˜ˆì‹œ 1: ë‹¨ì¼ ì„¹ì…˜ (ë„¤ì´ë²„ ê¸ˆìœµ ì£¼ì‹ ì‹œì„¸) - Blueprint 1ê°œ, entry_urls 1ê°œ
sample_single = NavigatorBlueprintCollection(
    total_jobs=1,
    blueprints=[
        NavigatorBlueprint(
            entry_urls=["https://finance.naver.com/sise/sise_rise.nhn"],
            total_layers=1,
            layers=[
                PageLayer(
                    layer_name="ì£¼ì‹ ìƒìŠ¹ ì¢…ëª© í…Œì´ë¸”",
                    url_pattern="https://finance.naver.com/sise/sise_rise.nhn",
                    selectors={"stock_name": "td.col_type1 a", "change_rate": "td.rate_t"},
                    navigate_to_next=None,
                    pagination_method="URLíŒŒë¼ë¯¸í„°"
                )
            ],
            rendering_type="Static SSR",
            anti_bot_notes="ì—†ìŒ"
        )
    ]
)

# ì˜ˆì‹œ 2: ë™ì¼ êµ¬ì¡°, ì¶œë°œ URLë§Œ ë‹¤ë¥¸ ê²½ìš° (ì •ì¹˜ë©´ + ì‚¬íšŒë©´) - Blueprint 1ê°œ, entry_urls 2ê°œ
sample_multi_url = NavigatorBlueprintCollection(
    total_jobs=1,
    blueprints=[
        NavigatorBlueprint(
            entry_urls=[
                "https://news.naver.com/section/100",  # ì •ì¹˜
                "https://news.naver.com/section/102",  # ì‚¬íšŒ
            ],
            total_layers=2,
            layers=[
                PageLayer(
                    layer_name="ë‰´ìŠ¤ ëª©ë¡",
                    url_pattern="https://news.naver.com/section/{sid}",
                    selectors={"article_link": "a.sa_text_title"},
                    navigate_to_next="a.sa_text_title",
                    pagination_method="AJAXë²„íŠ¼"
                ),
                PageLayer(
                    layer_name="ê¸°ì‚¬ ìƒì„¸",
                    url_pattern="https://n.news.naver.com/mnews/article/{mid}/{aid}",
                    selectors={"title": "h2.media_end_head_headline", "body": "#dic_area", "datetime": ".media_end_head_info_dateline"},
                    navigate_to_next=None
                )
            ],
            rendering_type="Dynamic CSR/JS",
            anti_bot_notes="ë”ë³´ê¸° ë²„íŠ¼ AJAX ë°©ì‹"
        )
    ]
)

# ì˜ˆì‹œ 3: êµ¬ì¡°ê°€ ë‹¤ë¥¸ ë‘ ì‚¬ì´íŠ¸ - Blueprint 2ê°œ
sample_multi_blueprint = NavigatorBlueprintCollection(
    total_jobs=2,
    blueprints=[
        NavigatorBlueprint(
            entry_urls=["https://news.naver.com/section/100"],
            total_layers=2,
            layers=[
                PageLayer(layer_name="ë‰´ìŠ¤ ëª©ë¡", url_pattern="https://news.naver.com/section/100",
                          selectors={"article_link": "a.sa_text_title"}, navigate_to_next="a.sa_text_title"),
                PageLayer(layer_name="ê¸°ì‚¬ ìƒì„¸", url_pattern="https://n.news.naver.com/mnews/article/{mid}/{aid}",
                          selectors={"title": "h2.media_end_head_headline", "body": "#dic_area"}, navigate_to_next=None)
            ],
            rendering_type="Dynamic CSR/JS", anti_bot_notes="ì—†ìŒ"
        ),
        NavigatorBlueprint(
            entry_urls=["https://www.danawa.com/product/?cateCode=112758"],
            total_layers=2,
            layers=[
                PageLayer(layer_name="ìƒí’ˆ ëª©ë¡", url_pattern="https://www.danawa.com/product/?cateCode=112758",
                          selectors={"product_link": "a.prod_name"}, navigate_to_next="a.prod_name", pagination_method="URLíŒŒë¼ë¯¸í„°"),
                PageLayer(layer_name="ìƒí’ˆ ìƒì„¸", url_pattern="https://prod.danawa.com/info/?pcode={id}",
                          selectors={"name": "h3.prod_tit", "price": "span.low_price", "specs": "div.spec_list"}, navigate_to_next=None)
            ],
            rendering_type="Dynamic CSR/JS", anti_bot_notes="ì—†ìŒ"
        )
    ]
)
````
